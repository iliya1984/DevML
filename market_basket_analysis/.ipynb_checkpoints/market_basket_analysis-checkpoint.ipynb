```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import networkx as nx
import warnings

#from mlxtend.frequent_patterns import apriori
#from mlxtend.frequent_patterns import association_rules
```

```python
sns.set(style="darkgrid", color_codes=True)
pd.set_option('display.max_columns', 75)
```

## Dataset

```python
data_df = pd.read_csv('../datasets/market_basket/market_basket.csv', header=None)
data_df.head()
```

```python
data_df.info()
```

```python
data_df.describe()
```

```python
bins = np.linspace(0, 1, 40)
color = plt.cm.rainbow(bins)
food_frequencies = data_df[0].value_counts().head(40)

food_frequencies.plot.bar(color=color,figsize=(13, 5))

plt.title('Frequency of the most popular items', fontsize=20)
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
data_df['food'] = 'food'
food = data_df.truncate(before = -1, after = 15)
food = nx.from_pandas_edgelist(food, source = 'food', target=0, edge_attr = True)
```

```python
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (13, 13)
pos = nx.spring_layout(food)

bins = np.linspace(0, 15, 1)
color = plt.cm.Set1(bins)

nx.draw_networkx_nodes(food, pos, node_size=12000, node_color = color)
nx.draw_networkx_edges(food, pos, width = 3, alpha = 0.6, edge_color='black')
nx.draw_networkx_labels(food, pos, font_size=15, font_family='sans-serif')

plt.axis('off')
plt.grid()
plt.title('Top 15 first choices', fontsize = 20)
plt.show()
```
