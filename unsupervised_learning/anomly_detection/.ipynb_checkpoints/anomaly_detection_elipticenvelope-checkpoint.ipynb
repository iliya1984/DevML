## Helper classes

```python
class CategoryItems:
    
    def __init__(self, items):
        self.items = items
        self.normal_items = None
        self.anomaly_items = None
        self.df = None

class Category:
    
    def __init__(self, category_id, name):
        self.category_id = category_id
        self.category_name = name
        self.items = None
        
    def hist(self, ax):
        self.items.items.hist(ax=ax,bins=32)
        
        
class CategoryDict(dict):
    
    def get_value(self, category_id):
        return self[f'class_{category_id}']
    
    def set_value(self, category_id, value):
        
        category_name = f'class_{category_id}'
        category = Category(category_id, category_name)
        
        items = CategoryItems(value)
        category.items = items
        
        self[category_name] = category
        
class EllipticEnvelopeModel:
    
    def __init__(self, contamination):
        self._model = EllipticEnvelope(contamination = contamination)
        
    def fit_transform(self, series):
        train_df = series.values.reshape(-1,1)
        self._model.fit(train_df)
        df = pd.DataFrame(series)
        df['deviation'] = self._model.decision_function(train_df)
        df['anomaly'] = self._model.predict(train_df)
        
        return df
        
```

```python
# creation of 4 differents data set based on categories defined before
category_dict = CategoryDict()

index = 0
for category in category_list:
    value = data_df.loc[ \
        data_df[ATSFDataset.Columns.CATEGORIES] == category, ATSFDataset.Columns.VALUE_C]
    
    category_dict.set_value(index, value)
    index = index + 1
```

```python
# plot the temperature repartition by categories
fig, axs = plt.subplots(2,2)
category_dict.get_value(0).hist(ax=axs[0,0])
category_dict.get_value(1).hist(ax=axs[0,1])
category_dict.get_value(2).hist(ax=axs[1,0])
category_dict.get_value(3).hist(ax=axs[1,1])
```

## Apply Eliptic Envelope algorithm on each category

```python
outliers_fraction = 0.01
```

```python
category =  category_dict.get_value(0)
category.category_name
```

```python
#Category training example
x_train = category.items.items.values.reshape(-1,1)
envelop_model = EllipticEnvelope(contamination = outliers_fraction)
envelop_model.fit(x_train)

analysis_df = pd.DataFrame(category.items.items)
analysis_df['deviation'] = envelop_model.decision_function(x_train)
analysis_df['anomaly'] = envelop_model.predict(x_train)
analysis_df.head()
```

```python
for category_name in category_dict:
    category = category_dict[category_name]
    
    model = EllipticEnvelopeModel(contamination=outliers_fraction)
    category_df = model.fit_transform(category.items.items)

    category.items.normal_items = category_df.loc[category_df['anomaly'] == 1, ATSFDataset.Columns.VALUE_C]
    category.items.anomaly_items = category_df.loc[category_df['anomaly'] == -1, ATSFDataset.Columns.VALUE_C]
    category.items.df = category_df
```

```python
category =  category_dict.get_value(0)
category.items.df.head()
```

```python
a0 = category_dict.get_value(0).items.normal_items
b0 = category_dict.get_value(0).items.anomaly_items

a1 = category_dict.get_value(1).items.normal_items
b1 = category_dict.get_value(1).items.anomaly_items

a2 = category_dict.get_value(2).items.normal_items
b2 = category_dict.get_value(2).items.anomaly_items

a3 = category_dict.get_value(3).items.normal_items
b3 = category_dict.get_value(3).items.anomaly_items

fig, axs = plt.subplots(2,2)
axs[0,0].hist([a0,b0], bins=32, stacked=True, color=['blue', 'red'], label=['normal', 'anomaly'])
axs[0,1].hist([a1,b1], bins=32, stacked=True, color=['blue', 'red'], label=['normal', 'anomaly'])
axs[1,0].hist([a2,b2], bins=32, stacked=True, color=['blue', 'red'], label=['normal', 'anomaly'])
axs[1,1].hist([a3,b3], bins=32, stacked=True, color=['blue', 'red'], label=['normal', 'anomaly'])
axs[0,0].set_title("WeekEndNight")
axs[0,1].set_title("WeekEndLight")
axs[1,0].set_title("WeekDayNight")
axs[1,1].set_title("WeekDayLight")
plt.legend()
plt.show()
```

```python
category_dfs = []
for category_name in category_dict:
    category = category_dict[category_name]
    category_dfs.append(category.items.df)

all_category_df = pd.concat(category_dfs)
all_category_df.head()
```

```python
data_df['anomaly'] = all_category_df['anomaly']
data_df['anomaly'] = np.array(data_df['anomaly'] == -1).astype(int)
```
