```python
class Statistics():
    
    def __init__(self):
        self.max_clusters = None
        self.max_score = None
        self.min_score = None
        self.score_range = None
        
    def info(self):
        print('K means algorithm run statistic:\n')
        print(f'Max clusters={self.max_clusters}')
        print(f'Max score={self.max_score}')
        print(f'Min score={self.min_score}')
        print(f'Score range={self.score_range}')

class KMeansConfiguration:
    
    def __init__(self, max_clusters):
        self.max_clusters = max_clusters

class ScoresDFColumns:
    N_CLUSTER = 'n_cluster'
    SCORE = 'score'
        
class KMeansAlgorithm:
    _configuration = None
    _kmeans = []
    
    def __init__(self, configuration):
        self._configuration = configuration
        
    def fit(self, df):
        max_clusters = self._configuration.max_clusters
        n_clusters_range = range(1, max_clusters)
        self._kmeans = [KMeans(n_clusters=i, n_init='auto').fit(df) for i in n_clusters_range]
        
        scores_range = range(len(self._kmeans))
        scores_tuples = [(i, self._kmeans[i].score(df)) for i in scores_range]

        scores_df = pd.DataFrame(columns=[ScoresDFColumns.N_CLUSTER, ScoresDFColumns.SCORE])
        scores_df[ScoresDFColumns.N_CLUSTER] = [t[0] for t in scores_tuples]
        scores_df[ScoresDFColumns.SCORE] = [t[1] for t in scores_tuples]
        
        self._scores = scores_df
        
    def get_scores(self):
        return self._scores
        
    def get_kmeans(self, n_clusters):
        return self._kmeans[n_clusters]
    
    def get_statistics(self):
        max_score = self._scores[ScoresDFColumns.SCORE].max()
        min_score = self._scores[ScoresDFColumns.SCORE].min()
        
        statistics = Statistics()
        statistics.max_clusters = self._configuration.max_clusters
        statistics.max_score = max_score
        statistics.min_score = min_score
        statistics.score_range = max_score - min_score
        
        return statistics
    
    def plot_scores(self):
        df = self._scores
        
        x = df[ScoresDFColumns.N_CLUSTER]
        y = df[ScoresDFColumns.SCORE]
        
        fig, ax = plt.subplots()
        ax.plot(x, y)
        plt.show()
```

```python
config = KMeansConfiguration(max_clusters=20)

model = KMeansAlgorithm(config)
model.fit(train_df)

model.plot_scores()
```

```python
statistics = model.get_statistics()
statistics.info()
```

```python
# Not clear for me, I choose 15 centroids arbitrarily and add these data to the central dataframe
cluster_count = 14
kmeans_results_df = pd.DataFrame(columns=['cluster', 'feature1', 'feature2'])

kmeans_results_df['cluster'] = kmeans[cluster_count].predict(train_df)
kmeans_results_df['feature1'] = train_df[0]
kmeans_results_df['feature2'] = train_df[1]
kmeans_results_df.head()
```

```python
def get_distance_by_point(df, model):
    
    df_len = len(df)
    distance = pd.Series()
    for i in range(0, df_len):
        point_coord = np.array(df.loc[i])
        centroid_coord = model.cluster_centers_[model.labels_[i]-1]
        distance.at[i] = np.linalg.norm(point_coord-centroid_coord)
    return distance
```

```python
outliers_fraction = 0.01
# get the distance between each point and its nearest centroid. The biggest distances are considered as anomaly
distance = get_distance_by_point(train_df, kmeans[14])
number_of_outliers = int(outliers_fraction * len(distance))
threshold = distance.nlargest(number_of_outliers).min()
kmeans_results_df['is_anomaly'] = (distance >= threshold).astype(int)

#kmeans_results_df[kmeans_results_df['anomaly'] == 1]
kmeans_results_df.head()
```

```python
# visualisation of anomaly with cluster view
fig, ax = plt.subplots()
colors = {0:'blue', 1:'red'}
ax.scatter(kmeans_results_df['feature1'], \
           kmeans_results_df['feature2'], \
           c=kmeans_results_df["is_anomaly"].apply(lambda x: colors[x]))
plt.show()
```

```python
# visualisation of anomaly throughout time (viz 1)
fig, ax = plt.subplots()

a = data_df.loc[kmeans_results_df['is_anomaly'] == 1, ['time_epoch', 'value_c']] #anomaly

ax.plot(data_df['time_epoch'], data_df['value_c'], color='blue')
ax.scatter(a['time_epoch'],a['value_c'], color='red')
plt.show()
```

```python
# visualisation of anomaly with temperature repartition (viz 2)
a = data_df.loc[kmeans_results_df['is_anomaly'] == 0, 'value_c']
b = data_df.loc[kmeans_results_df['is_anomaly'] == 1, 'value_c']

fig, axs = plt.subplots()
axs.hist([a,b], bins=32, stacked=True, color=['blue', 'red'], label=['normal', 'anomaly'])
plt.legend()
plt.show()
```
